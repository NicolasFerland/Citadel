# Merge

```python
community_health_pivot = pd.pivot_table(community_health[(
    community_health['indicator']=='Salmonella incidence per 100,000')|(
    community_health['indicator']=='E. coli O157 incidence per 100,000')|(
    community_health['indicator']=='Shigella incidence per 100,000')], index='county',columns='indicator')[['event_count','precent_or_rate']]
community_health_pivot
```

```python
food_establishment_inspections_county = food_establishment_inspections.groupby('county')['num_critical','num_noncritical','num_critical_not_corrected'].sum()
food_establishment_inspections_county
```

```python
demographics_county = demographics.groupby('county').mean()
demographics_county
```

```python
merge_df = food_establishment_inspections_county.join(community_health_pivot['precent_or_rate']).join(demographics_county)

features_bypopulation = ['num_critical','num_noncritical','total_households','$9,999_or_less', '$10,000_to_$14,999',
       '$15,000_to_$24,999', '$25,000_to_$34,999', '$35,000_to_$49,999',
       '$50,000_to_$74,999', '$75,000_to_$99,999', '$100,000_to_$150,000',
       '$150,000_to_$199,999', '$200,000_or_more','food_stamp_benefits', 'pop_w_health_insurance','pop_w_private_health', 'pop_w_public_coverage',
       'pop_no_health_insurance'
       ]
merge_df[features_bypopulation] = merge_df[features_bypopulation].div(demographics_county['population'], axis='index')

merge_df
```

```python
merge_df.to_csv('merged.csv')
```
